```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as px
```

```python
wec_df = pd.read_csv('./World Energy Consumption.csv')
wec_df
```

```python
countries = ['India','Russia','China','United States','Australia','Canada','Afghanistan','Africa',
             'Germany','Europe','Japan','Saudi Arabia','Ukraine','Brazil','Spain',
             'United Arab Emirates','United Kingdom', 'South Korea','Turkey','Taiwan'
            ]
```

```python
filt = wec_df["country"].isin(countries)
wec_df = wec_df[filt]
```

```python
wec_df
```

```python
wec_df = wec_df[wec_df['year'] >= 2000]
wec_df = wec_df[wec_df['year'] <= 2020]
```

```python
wec_df.isna().sum()
```

```python
wec_df.fillna(method = 'bfill',inplace = True)
wec_df.fillna(method = 'ffill',inplace = True)
```

```python
wec_df
```

```python
columns = [ # country Info
            'country','year', 'population','gdp',
            
            # Electricity Generation from various energy sources
            'oil_electricity','coal_electricity', 'gas_electricity', 
            'hydro_electricity', 'nuclear_electricity','wind_electricity',
            'biofuel_electricity', 'solar_electricity',
            'fossil_electricity','renewables_electricity','electricity_generation',
    
            # Share of  electricity consumption from various energy sources
            'coal_share_elec','oil_share_elec','gas_share_elec',
            'solar_share_elec', 'wind_share_elec','hydro_share_elec',
            'biofuel_share_elec','nuclear_share_elec',
            'fossil_share_elec','renewables_share_elec',

            # per capita of each energy source
            'biofuel_elec_per_capita', 'hydro_elec_per_capita', 'nuclear_elec_per_capita',
            'wind_elec_per_capita', 'coal_elec_per_capita','gas_elec_per_capita', 'oil_elec_per_capita',
            'solar_elec_per_capita','per_capita_electricity',

            # Energy consumption from various sources
            'solar_consumption', 'wind_consumption',
            'oil_consumption', 'nuclear_consumption','hydro_consumption',
            'gas_consumption','coal_consumption', 'biofuel_consumption',
            'primary_energy_consumption','renewables_consumption'

 ]
```

```python
wec_df = wec_df[columns]
```
